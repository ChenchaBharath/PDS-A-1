# creating new column Average

```python
data_frame["AverageScore"]=(data_frame["math score"]+data_frame["reading score"]+data_frame["writing score"])/3
print(data_frame["AverageScore"])
```

```python

def Grade(avg):
    if (avg >= 90):
        return 'A'
    elif (avg>=80 and avg<90):
        return 'B'
    elif (avg>=85 and avg<90):
        return 'B'
    elif (avg>=70 and avg<80):
        return 'C'
    elif (avg>=55 and avg<70):
        return 'D'
    else:
        return 'F'
    
data_frame["grade"] = data_frame.apply(lambda x : Grade(x["AverageScore"]),axis=1)
    
```

```python
# using Seaborn for analysis
```

```python
import seaborn as sns
import numpy as np
```

### below countplot describes the grade got by students grouped by their gender.It shows mainly how the student performance is effected by each category i.e gender.

```python
p = sns.countplot(y="grade", hue="gender", data=data_frame, order=[ "A", "B", "C", "D", "F"])
img = p.get_figure()
img.savefig("C:/Users/bhara/OneDrive/Desktop/Principles-of-DS/studentperformance/results/Analysis_Grade_Gender.png")
```
